Based on the subset above, the columns '**DateofHire', 'DateofTermination', and 'LastPerformanceReview_Date'** all follow the **date format 'M/DD/YYYY**'. In contrast, the **'DOB' column** uses a different format, **'MM/DD/YY'**. Consequently, the preprocessing of these columns will need to be handled separately to account for their differing date structures.

```python
# Create a date parser
def date_parser(date_str):
    try:
        date = datetime.strptime(date_str, '%m/%d/%y')

        if date.year > datetime.now().year:
            return date.replace(year=date.year - 100)
        return date
    except ValueError:
        return pd.NaT
```

```python
# Handle 'DOB' column (format: MM/DD/YY), data type conversion str -> datetime
df['DOB'] = pd.to_datetime(df['DOB'].astype(str).apply(date_parser), errors='coerce')
```

```python
# Check the result from the data conversion
df.DOB.head(10)
```

```python
# Convert all other date columns from data type str to datetime
date_columns = ['DateofHire', 'DateofTermination', 'LastPerformanceReview_Date']

for col in date_columns:
    df[col] = pd.to_datetime(df[col], format='%m/%d/%Y', errors='coerce')
```

```python
# Check results from data conversion
df[date_columns].head(10)
```

```python
# Check the whole dataset
df.info()
```

###**Feature Engineering**

```python
current_date = datetime.now()
```

```python
print("Current date used for calculation:", current_date)
```

```python
# Create new columns
df['Age'] = ((current_date - df['DOB']).dt.days / 365.25).apply(np.floor).astype(int)
df['AgeAtHiring'] = ((df['DateofHire'] - df['DOB']).dt.days / 365.25).apply(np.floor).astype(int)
df['Tenure'] = ((df['DateofTermination'].fillna(current_date) - df['DateofHire']).dt.days / 365.25).round().astype(int)
df['YearsSinceLastReview'] = ((current_date - df['LastPerformanceReview_Date']).dt.days / 365.25).round().astype(int)
```

# **Exploratory Data Analysis (EDA)**


```python
# Check numerical columns for outliers
numerical_columns = ['Salary', 'EngagementSurvey', 'EmpSatisfaction', 'SpecialProjectsCount', 'DaysLateLast30', 'Absences']

plt.figure(figsize=(12, 8))

for i, col in enumerate(numerical_columns, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(data=df, x=col)
    plt.title(f'Box Plot of {col}')

plt.tight_layout()
plt.show()
```

**Salary:** The salary outliers include a range of high values, with notable entries such as 170,500, 180,000, and even 250,000. These figures significantly exceed the typical salary range, indicating that there are a few employees with exceptionally high salaries.

**Engagement Survey:** Outliers in the Engagement Survey scores are notably low, with values such as 1.12 and 1.93. These scores are much lower than the typical range, suggesting that some employees may have very low engagement levels.

**Employee Satisfaction:** There are no outliers in the Employee Satisfaction data, implying that the satisfaction scores fall within a consistent range.

**Special Projects Count:** Outliers in the Special Projects Count include several high values, with counts up to 8. This suggests that some employees are involved in a notably higher number of special projects compared to others.

**Days Late Last 30:** The outliers for Days Late Last 30 include several instances of values from 2 to 6. These represent the higher end of lateness in the past month, indicating occasional but not extreme lateness.

**Absences:** There are no outliers in the Absences data, suggesting that the number of absences is uniformly distributed within the expected range.
